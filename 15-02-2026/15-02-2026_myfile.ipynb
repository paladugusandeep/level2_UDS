## Creating numpy arrays    

```python
import numpy as np
```

```python
# list1=[1,2,3,4] 1 dimensional
arr_1=np.array([1,2,3,4])
print(arr_1)
print(type(arr_1))
```

```python
arr_1
```

```python
arr_2= np.array([[1,2,3],[4,5,6]])  # 2 dimensional
print(arr_2)
```

```python
# Understanding of datatype
arr_3=np.array([1,2,3,4],dtype=np.int8)
print(type(arr_3))
```

```python
# To know the type of array values
type(arr_3[0])
```

## np.linspace(start,stop,num=x)

```python
print(np.linspace(1,100,4))
```

```python
print(np.linspace(1,100,4,dtype=np.int8))
```

```python
a_linspace=np.array(np.linspace(1,10,10))
print(a_linspace)
```

```python
for num in a_linspace:
    print(num)
    print(type(num))
```
