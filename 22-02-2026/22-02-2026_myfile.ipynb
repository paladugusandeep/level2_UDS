# Slicing Rows and columns together

```python
# Rows 0 to 4 , only name and age
df[0:5][['Name','Age']]
```

```python
# Another way to achieve above problem statement
df.loc[0:4,['Name','Age']]
```

```python
# One more way to achieve above problem statement
df.loc[:4,['Name','Age']]
```

```python
#First 3 rows and first 2 columns
df.iloc[0:3,0:2]
```

```python
# First 3 rows and COlumn Name as first and then Survived
df.loc[0:3,['Name','Survived']]
```

```python
# Same problem statement using iloc
df.iloc[0:3,[3,1]]
```
