## Creating a dataframe

```python
import pandas as pd
```

```python
data={
    "Name":['Sandeep','Padmini','Shannu'],
    "Age":['39','37','6'],
    "City":['GDK','HYT','HYD']
}
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
df
```

```python
# From list of dictionaries
records=[
    {'Name':'Sandeep','Age':39,'City':'GDK'},
    {'Name':'Padmini','Age':37,'City':'HYT'},
    {'Name':'Shannu','Age':6,'City':'HYD'}
]
df=pd.DataFrame(records)
df
```

```python
import numpy as np
# From numpy arrays
arr1=np.array([[1,2],[3,4],[5,6]])
arr1
df=pd.DataFrame(arr1,columns=['A','B'])
df
```

```python
data={
    "Name":['Sandeep','Padmini','Shannu'],
    "Age":['39','37','6'],
    "City":['GDK','HYT','HYD']
}
df=pd.DataFrame(data)
df
```

```python
# Accessing the values
print(df['Name']) # Series (a column)
print('============')
print(df[['Name','Age']]) # Dataframe with 2 columns
```
