# What is Pandas?

Pandas is an open-source Python library for data manipulation and analysis. It gives you 
powerful, fast, flexible, and expressive data structures designed to work with structured data.

It is built on top of NumPy, and its name comes from â€œPanel Dataâ€ (econometrics term for multidimensional data).

# Why Pandas? What problem does it solve?

### Before Pandas:

People used raw Python lists, dictionaries, or NumPy arrays.

These structures lacked labels, indexing, grouping, join/merge, handling missing data, etc.

Real-world datasets (CSV, JSON, Excel, SQL) are structured in rows and columns â€” not arrays.

So Pandas was designed to work like:

ðŸ“Š Excel + ðŸ§® NumPy + ðŸ—‚ï¸ SQL + ðŸ§° ETL tools â€” inside Python

```python
# Pandas resolves s2[0] in the following order:

# Check if 0 is a label in the index
# â†’ If yes, return by label

# If not, and if the index is not integer-based, then treat 0 as a position
# â†’ Return item at index position 0 (first element)

# FutureWarning: Series.__getitem__ treating keys as positions is deprecated. 
# In a future version, integer keys will always be treated as 
# labels (consistent with DataFrame behavior). 
# To access a value by position, use `ser.iloc[pos]`

print(s2[0])
```

# 2. What is a DataFrame?

A DataFrame is a 2D labeled data structure with columns of potentially different types, just like a spreadsheet or SQL table.

Think of it like:

- A dict of Series

- Or a table with rows and columns

It is the core structure of pandas used for:

- Reading, cleaning, transforming, analyzing, and exporting data


### Why DataFrame?

Real-world data is usually tabular (e.g., CSV, Excel, SQL)

Rows = records

Columns = features

It supports label-based indexing, automatic alignment, missing data handling, grouping, aggregation, and more


| Concept     | Real-Life Analogy             |
| ----------- | ----------------------------- |
| `DataFrame` | Excel sheet or database table |
| `Series`    | One column of that sheet      |
| Index       | Row number or ID              |
| Columns     | Named headers                 |


```python
# From a dictionary of lists

data = {
    'Name' : ['Hemanth', 'Aditya', 'rajbeer'],
    'Age' : [30, 31, 32],
    'City' : ['Delhi', 'Mumbai', 'Bangalore']
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df
```

```python
# From a list of dicts

records = [
    {'Name' : 'Hemanth', 'Age' : 30, 'City' : 'Delhi'},
    {'Name' : 'Aditya', 'Age' : 31, 'City' : 'Mumbai'},
    {'Name' : 'Rajbeer', 'Age' : 32, 'City' : 'Bangalore'}
]

df = pd.DataFrame(records)
print(df)
```

```python
# From NumPy arrays (if all columns are of the same type)

arr = np.array([[1, 2], [3, 4], [5, 6]])
df = pd.DataFrame(arr, columns=['A', 'B'])
print(df)
```

```python
# Accessing Data

data = {
    'Name' : ['Hemanth', 'Aditya', 'rajbeer'],
    'Age' : [30, 31, 32],
    'City' : ['Delhi', 'Mumbai', 'Bangalore']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
print(df['Name']) # Series (a column)
print("===========")
print(df[['Name', 'Age']]) # DataFrame with two columns
print("===========")
print(df.iloc[0]) # Row by integer index label or position
print("===========")
print(df.loc[0])  # Row by index label
```

# 5. Conditional Selection (Boolean Indexing)

```python
# Passengers older than 70

df['Age'] > 70 # Boolean index of passengers > 70 (True) else False
```

```python
df[df['Age'] > 70]
```

```python
# Females who survived
# Sex = female AND survived = 1

df[(df['Sex'] == 'female') & (df['Survived'] == 1)] # --> True / False
```

```python
# get Name & Age of Females who survived
# df[["Name" , 'Age']]
female_survivors_subset = df[(df['Sex'] == 'female') & (df['Survived'] == 1)][['Name', 'Age']]
female_survivors_subset.head()
```

```python
# get Name & Age of Females who survived using loc

# df.loc[label, ['Name', 'Age']]

df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1), ['Name', 'Age']]
```
