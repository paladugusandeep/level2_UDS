# Iteration in Pandas

In pandas, iteration refers to looping over rows or columns of a DataFrame or Series. While pandas is optimized for vectorized operations, sometimes we need to loop — for example, for conditional logic, debugging, or generating complex values.

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Delhi', 'Mumbai', 'Chennai']
})
```

```python
df
```

```python
for index,row in  df.iterrows():
    print(f"{index: } Name: {row.Name} Age is: {row.Age} City: {row.City}")
```

```python
for row in  df.itertuples():
    print(f"{index: } Name: {row.Name} Age is: {row.Age} City: {row.City}")
```

```python
# Iterating Over Columns

for col in df.columns:
    print(f"Column: {col}")
    print(df[col].values, end='\n\n')
   
```

# Conditional Example: Add Column with Custom Logic

Based on age column, i want to create a new column that will classify people as 'young' or 'senior'

```python
categories = []

# Loops

for idx, row in df.iterrows():
    print(row)
    print(row['Age'])

    if row['Age'] < 40 :
        categories.append("young")
    else :
        categories.append("senior")

df['Category'] = categories

df.head()
```

# Better Way: Vectorized / Apply


```python
df['Category_vectorized_1'] = df['Age'].apply(lambda age : 'young' if age < 40 else 'senior')

df['Category_vectorized_1'].head()
```

```python
categorize_age_lambda = lambda age : 'young' if age < 40 else 'senior'

df['Category_vectorized_2'] = df['Age'].apply(categorize_age_lambda)

df['Category_vectorized_2'].head()
```

```python
def categorize_age_function(age):
    # if age < 40:
    #     return 'young'
    # else :
    #     return 'senior'
    return 'young' if age < 40 else 'senior'

df['Category_vectorized_3'] = df['Age'].apply(categorize_age_function)

df['Category_vectorized_3'].head()
```
