## Ravel    

```python

import numpy as np
```

```python
a=np.array([[1,2,3],[4,5,6]])
print(a)
```

```python
a_ravel=a.ravel()
print(a_ravel)
print(a.ravel())
```

```python
print(id(a))
print(id(a.ravel()))
print(id(a_ravel))
```

```python
print(f"Data address of 'a':  {a.__array_interface__['data'][0]}")
```

```python
print(f"Data address of 'a.ravel()':  {a.ravel().__array_interface__['data'][0]}")
```

```python
print(f"Data address of 'a_ravel: ' {a_ravel.__array_interface__['data'][0]}")
```
